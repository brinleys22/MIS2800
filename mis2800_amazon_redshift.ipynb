## YOUR ANSWERS TO PART 1 BELOW

Question 1: By moving to a data warehouse like Amazon Redshift through Amazon Web Services, Lyft was able to do more than just store ride data — they could analyze massive amounts of historical and real-time information together. Instead of simply tracking completed rides, they could identify patterns in rider demand, routes, pricing, and driver behavior. This deeper analysis made it possible to develop new, data-driven products like Lyft Line, which depends on matching riders efficiently based on direction, timing, and cost optimization.


Question 2: Lyft cannot use the same database for both app operations (OLTP) and strategic planning (OLAP) because they are designed for different purposes. The app requires fast, real-time transactions like booking rides and updating locations, which need quick reads and writes. In contrast, strategic planning involves complex queries across huge datasets to analyze trends and performance. Running heavy analytical queries on a transactional database would slow down the app, while a data warehouse is optimized for large-scale analysis but not rapid, real-time updates.

## YOUR ANSWERS TO PART 2 BELOW

Question 1: In Amazon Redshift, columnar storage is often more important than perfect normalization because analysts usually query a few columns across millions of rows. Columnar storage allows Redshift to read only the needed columns instead of entire rows, which reduces disk usage and speeds up queries. For analytics workloads, this design improves performance more than strictly normalized tables would.

Question 2: Redshift uses a Leader Node and multiple Compute Nodes to speed up reporting. The Leader Node receives and plans the query, then distributes the work across Compute Nodes that process data in parallel. This parallel architecture allows complex Business Intelligence reports to be generated much faster than if everything ran on a single machine.
